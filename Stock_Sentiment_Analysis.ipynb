## Problem Statements: Stock Sentiment Analysis using News Headlines


*   Predict Stock Price Movement Based On News Headline using NLP



image.png

### Step1: Importing Library

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

### Step2: Load Data_Set

```python
df=pd.read_csv('kaggle_data_set.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
# split our data in train and test by date
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

### Step 3: Clean Data_Set

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)  # Now take all Top news channel removing all thing apart from [a-zA-Z]

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
# make paragraph from all 25 top news on evry row(date) and using this predict stock price would be increased or not on this perticular date
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
# make every row's top news as a paragraph
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

### Step 4: Converting into BagofWords and Implement RandomForest

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))  # take 2-2 word
traindataset=countvector.fit_transform(headlines)
```

```python
# implement RandomForest Classifier (Bagged of Decision Tree)
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

### Step 5: Prediction on Test Data

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

### Step 6: Accuracy and Confusion_Matrix

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
